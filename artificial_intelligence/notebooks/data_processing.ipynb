Importations

```python
import pandas as pd
import os
```

```python
def summary(dataframe):
    print(dataframe.head())
    print(dataframe.describe())

def null_checker(dataframe):
    print(dataframe.isnull().sum())

def duplicated_checker(dataframe):
    print(dataframe.loc[dataframe.duplicated(),:])
    # for column in list(dataframe.columns.values):
    #     print("Column: " + column)
        
```

```python
df = pd.read_csv("../data/raw/goafrica_entreprises.csv")
print("------------Summary------------")
summary(df)
```

```python
print("------------Null_checker------------")
null_checker(df)
```

```python
print("------------Duplicated_checker------------")
duplicated_checker(df)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("------------Duplicated_checker------------")
duplicated_checker(df)
```

```python
df.fillna("", inplace=True)
# df[['tel']].fillna(0, inplace=True)
df["summary"] = df["category"] + df["description"]
df.head()
```

```python
print("------------Null_checker------------")
null_checker(df)
```

```python
df.drop(df[df['summary'] == ""].index, inplace = True)
```

```python
df.head()
```

```python
indices_to_drop = df[df['category'] == ''].index
df.drop(indices_to_drop, inplace=True)
df[df['category'] == '']
```
