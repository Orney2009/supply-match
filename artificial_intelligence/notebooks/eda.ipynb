```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def barplotter(title, dataset, column, labels):
    plt.rcParams["figure.figsize"] = [18,10]
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}

    plt.title(title, fontdict=font1)
    counts = dataset[column].value_counts().reset_index()
    counts.sort_values(by='count', inplace=True, ascending=False)

    counts = counts.groupby(['count']).size()

    print(counts)

    # plot = sns.barplot(x=counts[column], y=counts['count'])
    plot = sns.barplot(x=counts.index, y=counts.values, hue=counts.index, legend=False, palette='Set2')
    plt.xlabel(labels[0], fontdict=font2)
    plt.ylabel(labels[1], fontdict=font2)
    plt.show()

    fig = plot.get_figure()
    fig.savefig(f"../figures/{title}.png")


def top_sectors(dataset, column):    
    counts = dataset[column].value_counts().reset_index()
    counts.sort_values(by='count', inplace=True, ascending=False)
    counts = counts.head(n=20)
    print(counts)


def tail_sectors(dataset, column):    
    counts = dataset[column].value_counts().reset_index()
    counts.sort_values(by='count', inplace=True, ascending=False)
    counts = counts.tail(n=20)
    print(counts)
    
```

```python
df = pd.read_csv("../data/processed/processed_data.csv")
```

```python
print(f"Total number of entreprises: {df.shape[0]}")
```

```python

barplotter("Count of entreprises grouped by sector", df, "category", ['Number of entreprises', 'Number of sectors'])
```

```python
top_sectors( df, 'category')
```

```python
tail_sectors( df, 'category')
```
